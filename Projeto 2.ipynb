```python
import pandas as pd

p1 = pd.read_excel("Arquivo1.xlsx" )
p2 = pd.read_excel("Arquivo2.xlsx" )

display(p1)
display(p2)
```

```python
p3 = pd.merge(p1,p2, on = 'Game')
display(p3)
```

```python
#colunas em variáveis
metascore = p3['Metascore']
price = p3['Price']
gametype = p3[['Game_Type']]
game = p3[['Game']]
release_date = p3[['Release_date']]
owners = p3[['Owners']]
average = p3[['Average_Hours-Played_Since_2009']]
publishers = p3[['Publishers']]
currentplayers = p3[['Current_Players']]
peakplayers = p3[['Peak_Players']]
hoursplayed = p3[['Hours_Played']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Quais empresas produziram mais jogos
sns.countplot(data = p3, x = 'Publishers')
plt.show()
p3['Publishers'].value_counts()
```

```python
#Quais jogos foram produzidos pelas 4 principais empresas?
display(p3[['Game_Type','Game','Price','Publishers']].loc[p3['Publishers'] == 'Valve'])
display(p3[['Game_Type','Game','Price','Publishers']].loc[p3['Publishers'] == 'Studio Wildcard'])
display(p3[['Game_Type','Game','Price','Publishers']].loc[p3['Publishers'] == 'Klei Entertainment'])
display(p3[['Game_Type','Game','Price','Publishers']].loc[p3['Publishers'] == 'BetaDwarf'])

```

```python
#O jogo com mais hora jogadas
maishoras = max(p3['Hours_Played'])
display(p3.loc[p3['Hours_Played'] == maishoras])


#O jogo com mais jogadores no pico
jogadorespico = max(p3['Peak_Players'])
display(p3.loc[p3['Peak_Players'] == jogadorespico])

#O jogo com mais jogadores 
jogadores = max(p3['Current_Players'])
display(p3.loc[p3['Current_Players'] == jogadores])


#O jogo mais caro
preco = max(p3['Price'])
display(p3.loc[p3['Price'] == preco])
```

```python
import numpy as np


#Diferença em % de peak players para current players
currentplayers = np.array(p3[['Current_Players']])
peakplayers = np.array(p3[['Peak_Players']])

current_peak = peakplayers/currentplayers
p3['Current_Peak'] = current_peak
```

```python
#Quantos jogos gratuitos tem
#p3['Price'].value_counts()
gratuitos = p3.loc[p3['Price'] == 0.00]
pagos = p3.loc[p3['Price'] != 0.00]
print(len(gratuitos),"jogos gratuitos")
print(len(pagos),"jogos pagos")
```

```python
#10 jogos com maior diferença
display(p3[['Price','Game_Type','Game','Owners','Current_Players','Current_Peak']].sort_values(by = ['Current_Peak'], ascending = False)[:9])

sns.scatterplot(y= p3['Current_Players'],x = p3['Current_Peak'])
plt.show()
```

```python
#10 jogos com maior diferença pagos
display(pagos.sort_values(by = ['Current_Peak'], ascending = False)[:9])

sns.scatterplot(y= pagos['Current_Players'],x = pagos['Current_Peak'])
plt.show()
```

```python
#10 jogos com maior diferença gratuitos
display(gratuitos.sort_values(by = ['Current_Peak'], ascending = False)[:9])

sns.scatterplot(y= gratuitos['Current_Players'],x = gratuitos['Current_Peak'])
plt.show()
```

```python
#Preço por categoria
tipojogo= p3[['Game_Type','Price','Average_Hours-Played_Since_2009']].groupby(['Game_Type']).mean()
display(tipojogo)

#Preço por categoria de jogos pagos
tipojogopg= pagos[['Game_Type','Price','Average_Hours-Played_Since_2009']].groupby(['Game_Type']).mean()
display(tipojogopg)

p3['Game_Type'].value_counts()
```
